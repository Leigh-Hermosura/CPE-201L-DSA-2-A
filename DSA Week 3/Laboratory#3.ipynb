```python

testScores = [26, 49, 98, 87, 62, 75]
oddNum = []
evenNum = []
total = 0

for number in testScores:
  if number % 2 == 0: #even number checker
    evenNum.append(number)
    total += number

  elif number % 2 != 0: #odd number checker
    oddNum.append(number)

print(f"Odd numbers: {oddNum}")
print(f"""
Even numbers: {evenNum}
Even number total: {total}""")
total = 0
```
